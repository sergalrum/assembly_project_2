# Основные понятия и описание данных

Технологический процесс
- Rougher feed — исходное сырье
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
   - Xanthate — ксантогенат (промотер, или активатор флотации);
   - Sulphate — сульфат (на данном производстве сульфид натрия);
   - Depressant — депрессант (силикат натрия).
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота

Параметры этапов
- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи

# Наименование признаков

Наименование признаков должно быть такое:  
[этап].[тип_параметра].[название_параметра]

Пример: rougher.input.feed_ag  
Возможные значения для блока [этап]:
- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики

Возможные значения для блока [тип_параметра]:
- input — параметры сырья
- output — параметры продукта
- state — параметры, характеризующие текущее состояние этапа
- calculation — расчётные характеристики

### Проверим, что эффективность обогащения рассчитана правильно

Вычислим её на обучающей выборке для признака `rougher.output.recovery`.

Найдем MAE между нашими расчётами и значением признака.

- C — доля золота в концентрате после флотации/очистки;
- F — доля золота в сырье/концентрате до флотации/очистки;
- T — доля золота в отвальных хвостах после флотации/очистки.

```python
C = gi_train['rougher.output.concentrate_au']
F = gi_train['rougher.input.feed_au']
T = gi_train['rougher.output.tail_au']

recovery = (C * (F -T)) / (F * (C -T)) * 100
```

```python
mae_recovery = mean_absolute_error(gi_train['rougher.output.recovery'], recovery)
print(f'MAE между нашими расчётами и значением признака: {mae_recovery}')
```

#### Вывод по итогу проверки расчета эффективности обогащения\

Значение MAE между нашими расчётами и значением признака: 9.83758577644259e-15, практически равное нулю. Это свидетельствует о правильности расчета эффективности обогощения.

### Функция для вычисления итоговой sMAPE

```python
# Функция для вычисления smape
# a - фактическое значение, f - прогноз
def smape(a, f):
    smape = (1 / len(a)) * np.sum(2 * np.abs(f-a) / (np.abs(a) + np.abs(f))) * 100
    return smape
```

```python
# Функция для вычисления итоговой smape для ndarray
def f_smape(targ, pred):
    res = smape(targ, pred)
    f_smape = 0.25 * res[0] + 0.75 * res[1]
    return f_smape
```

```python
#  Функция для вычисления итоговой smape для пд
def f_smape_test(targ, pred):
    res0 = smape(targ.iloc[:, 0], pred.iloc[:, 0])
    res1 = smape(targ.iloc[:, 1], pred.iloc[:, 1])
    f_smape = 0.25 * res0 + 0.75 * res1
    return f_smape
```

```python
f_smape_scorer = make_scorer(f_smape, greater_is_better=False)
```

```python
%%time

best_result_forest = 10000
best_depth_forest = 0
best_est_forest = 0

for est in range(10, 101, 20):
    for depth in range (12, 16):
        gi_model_forest = RandomForestRegressor(random_state=RANDOM_STATE, n_estimators=est, max_depth=depth)
        scores = []
        scores = cross_val_score(gi_model_forest, gi_train_feature, gi_train_target, cv=5, scoring=f_smape_scorer)
        result = scores.mean()
        
        if result < best_result_forest:
            best_result_forest = result
            best_depth_forest = depth
            best_est_forest = est
```
